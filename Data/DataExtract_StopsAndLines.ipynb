```python
import json
import urllib
import pprint
```

```python
url_tubeStopList = 'https://raw.githubusercontent.com/domherriott/london_map/master/Data/tube_stop_list.json'
```

```python
dLines = {}
url_search = 'https://api.tfl.gov.uk/Line/Mode/tube'
response = urllib.request.urlopen(str(url_search))
search = json.loads(response.read())
for i in search:
    dLines[i['id']] = {}

dLines
```

```python
for i in dLines:
    print(i)
    url_search = 'https://api.tfl.gov.uk/Line/' + i +'/StopPoints'
    response = urllib.request.urlopen(str(url_search))
    search = json.loads(response.read())
    for j in search:
        searchName = j['commonName']
        stopId = j['id']
        dLines[i][searchName] = stopId
    

dLines
```

```python
dStops = {}

response = urllib.request.urlopen(url_tubeStopList)
tubeStopList = json.loads(response.read())

for i in tubeStopList:

    name = i['Name']
    
    
    apiSearchName = i['Name'].replace(' ', '%20')
    url_search = 'https://api.tfl.gov.uk/StopPoint/Search?query=' + apiSearchName
    response = urllib.request.urlopen(str(url_search))
    search = json.loads(response.read())
    find = search['matches'][0]
    
    if ('tube' in find['modes']):
        searchName = find['name']
        if (searchName[-20:] != ' Underground Station'):
            searchName = searchName + ' Underground Station'
            
        dStops[searchName] = {
            'ID':         find['id'],
            'modes':      find['modes'],
            'name':       name,
            'lat':        find['lat'],
            'lon':        find['lon'],
            'zone':       find['zone'],
            'lines':      []
        }            
        
dStops["King's Cross St. Pancras Underground Station"] = dStops.pop("King's Cross & St Pancras International Underground Station")
dStops['Hammersmith Underground Station']['lines'].append('piccadilly')
dStops['Hammersmith Underground Station']['lines'].append('district')
dStops['Hammersmith Underground Station']['lines'].append('circle')
dStops['Heathrow Terminals 1-2-3 Underground Station']['lines'].append('piccadilly')
dStops['']
        
for i in dStops:
    for j in dLines:
        if (i in dLines[j]):
            dStops[i]['lines'].append(j)
    
dStops
```

```python
with open('dLines.json', 'w') as f:
    json.dump(dLines, f)
```

```python
with open('dStops.json', 'w') as f:
    json.dump(dStops, f)
```
