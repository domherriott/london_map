```python
import json
import urllib
import pprint
```

```python
geoJson = 'https://raw.githubusercontent.com/oobrien/vis/master/tube/data/tfl_lines.json'
response = urllib.request.urlopen(str(geoJson))
search = json.loads(response.read())
search = search['features']
```

```python
#Piccadilly line
pic = []

for i in range(41, 92):
    for j in search[i]['geometry']['coordinates']:
        print(i, j[0], j[1])
        pic.append(j)
        
for i in range(95, 97):
    for j in search[i]['geometry']['coordinates']:
        print(i, j[0], j[1])
        pic.append(j)
        
for i in range(420, 430):
    for j in search[i]['geometry']['coordinates']:
        print(i, j[0], j[1])
        pic.append(j)
        
        
for i in range(499, 502):
    for j in search[i]['geometry']['coordinates']:
        print(i, j[0], j[1])
        pic.append(j)


#pic
```
